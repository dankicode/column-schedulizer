```python
import tkinter
import pandas as pd
import numpy as np
import csv
# import openpyxl
import math
```

```python
db_path = '17300-01_240213_dk_grav_steel-col-data.xlsx'

# column_db_story_data = pd.read_excel(db_path, sheet_name='StoryData')
# columns = ['StoryUID', 'Level', 'Story_Label', 'FloorHeight', 'CummulativeHeight']
# column_db_story_data = column_db_story_data[columns]

column_db_column_data = pd.read_excel(db_path, sheet_name='ColumnData')
columns = ['Story_Label', 'x1', 'y1', 'z1', 'x2', 'y2', 'z2', 'NodeI', 'NodeJ', 'Size', 'Grid_Label']
column_db_column_data = column_db_column_data[columns]
```

```python
with open('17300-01_240213_dk_Col-Design.csv', 'r') as cdf:
    reader = csv.reader(cdf)
    
    info = {}
    raw_data = []
    story = []
    grid_loc = []
    size = []
    pu = []
    for i, row in enumerate(reader):
#         print(i, row)
        if 'Story' in row[0]:
            story.append(row[0].split(' ')[-1])
            grid_loc.append(row[1].split(' ' )[-1])
            
            raw_data.append(row[0].split(' ')[-1])
            raw_data.append(row[1].split(' ' )[-1])
        if 'Fy' in row[0]:
            size.append(row[-1])
            
            raw_data.append(row[-1])
        if 'Pu (kip)' == row[0]:
            pu.append(math.ceil(float(row[2].strip())))
            
            raw_data.append(math.ceil(float(row[2].strip())))
#             print(math.ceil(float(row[2].strip())))

# print(len(story), len(grid_loc), len(size), len(pu))
print(raw_data)
```

```python
schedule = dict()
i = 0

while i < len(raw_data):
    key = f'{raw_data[i]}'
    
    if schedule.get(key) is None:
        schedule[key] = dict()
        
        schedule[key][raw_data[i+1]] = []
        schedule[key][raw_data[i+1]].append(raw_data[i+2])
        schedule[key][raw_data[i+1]].append(raw_data[i+3])

    else:
        schedule[key][raw_data[i+1]] = []
        schedule[key][raw_data[i+1]].append(raw_data[i+2])
        schedule[key][raw_data[i+1]].append(raw_data[i+3])
    i += 4
print(schedule)
```

```python
reformed_dict = {}
for outerKey, innerDict in schedule.items():
    for innerKey, values in innerDict.items():
        reformed_dict[(outerKey, innerKey)] = values
        
multiIndex_df = pd.DataFrame(reformed_dict)
multiIndex_df.rename(index={0: 'Size', 1:'Pu'}, inplace = True)
```

```python
multiIndex_df.loc[:, '300_MC']
```

```python
final = []

for story in reversed(column_db_column_data.Story_Label.unique()):
    story_df = multiIndex_df.loc[:, story]
    story_df['story'] = story
    final.append(story_df)
schedule_df = pd.concat(final)
```

```python
schedule_df = schedule_df.reindex(schedule_df.columns.sort_values(), axis = 1)
story = schedule_df.story
schedule_df = schedule_df.drop(columns=['story'])
schedule_df.insert(loc=0, column='story', value=story)
schedule_df
```

```python
schedule_df.to_excel('17300-01_dk_python_steel_grav_col_schedule.xlsx')
```

```python
## Base Plates
with open('17717_240703_grav_base_plate_designs.csv', 'r') as cdf:
    reader = csv.reader(cdf)
    
    info = {}
    raw_data = []
    grid_loc = []
    size_n = []
    Pa = []
    plate_thickness = []
    size = []
    
    for i, row in enumerate(reader):
#         print(i, row)
        if 'Base Plate level ' in row[0]:
            grid_loc.append(row[-1].split(' ' )[-1])
            
            raw_data.append(row[1].split(' ' )[-1])

        # square base plates
        if 'N (parallel to web) (in) .' in row[0]:
            size_n.append(row[-1] + ' x ' + row[-1])
            
            raw_data.append(row[-1] + ' x ' + row[-1])

#         if 'B (perpendicular to web) (in) .' in row[0]:
#             size_b.append(row[-1])

        if 'Plate Thickness (in) .' == row[0]:
            plate_thickness.append(row[-1].strip())
            
            raw_data.append(row[-1].strip())
            
        if 'Column Size .' in row[0]:
            size.append(row[-1])
            
            raw_data.append(row[-1])
            
        if 'Dead Load (kips) . ' in row[0]:
#             dead.append(float(row[-1].strip()))
            dead = float(row[-1].strip())
#             print(type(dead))
            
        if 'Live Load (kips) .' in row[0]:
#             live.append(float(row[-1].strip()))
            live = float(row[-1].strip())
#             print(type(live))
    
            raw_data.append(math.ceil(dead + live))


# print(len(grid_loc), len(size_n), len(size), len(Pa), len(plate_thickness))
print(raw_data)
```

```python
bp_schedule = dict()

i = 0
while i < len(raw_data):
    key = f'{raw_data[i]}'
    
    if bp_schedule.get(key) is None:
        bp_schedule[key] = dict()
        bp_schedule[key] = []
        bp_schedule[key].append(raw_data[i+1])
        bp_schedule[key].append(raw_data[i+2])
        bp_schedule[key].append(raw_data[i+3])
        bp_schedule[key].append(raw_data[i+4])

    
    else: 
        bp_schedule[key] = []
        bp_schedule[key].append(raw_data[i])
        bp_schedule[key].append(raw_data[i+1])
        bp_schedule[key].append(raw_data[i+2])
        bp_schedule[key].append(raw_data[i+3])
        bp_schedule[key].append(raw_data[i+4])
    i += 5
```
